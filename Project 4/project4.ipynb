```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import numpy as np
import statistics
from sklearn.metrics import roc_curve, auc, mean_squared_error
import pickle
from collections import Counter
from math import sqrt
```

```python
data = pd.read_csv('data/network_backup_dataset.csv')
```

```python
# Functions

def plot_workflows(data):
    workflow_data = []
    uniq_wf = set(pd.unique(data['Work-Flow-ID']))
    for wf in uniq_wf:
        workflow_data.append(data.loc[data['Work-Flow-ID'] == wf])

    plt.figure(figsize=(20,10))
    for df, wf in zip(workflow_data, uniq_wf):
        temp_dt = df.groupby(['Week #', 'Day of Week'])['Size of Backup (GB)'].sum()
        plt.plot(range(1,len(temp_dt) + 1), temp_dt, label=wf)
    plt.xlabel('Days')
    plt.ylabel('Backup Sizes')
    plt.title('Days versus Backup Sizes (%d days)' %len(temp_dt))
    plt.legend(loc="upper right")
    plt.show()
```

# Task 1: Plot Workflows

```python
twenty_data = data.loc[data['Week #'].isin([1,2,3])]
plot_workflows(twenty_data)
plot_workflows(data)
```
